```python
#modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read csvs
annual_data = pd.read_csv('../raw_data/annual_reports_nps.csv')
monthly_data = pd.read_csv('../raw_data/monthly_nps.csv')
data_2019=pd.read_csv('../raw_data/backcountry2019.csv')
annual_data.head()
```

```python
#filter the raw data with columns of choice
annual_data.columns
#choosing columns
filtered_data=annual_data[['ParkName','ParkType', 'Region', 'State', 'Year', 'Month','RecreationVisits', 
                           'RecreationHours','TentCampers', 'RVCampers', 'Backcountry','RecreationVisitsTotal',
                           'RecreationHoursTotal','TentCampersTotal', 'RVCampersTotal','BackcountryTotal']]
filtered_data
```

```python
#ok filtering columns again
filtered_data.columns
value_columns= ["RecreationVisits", 'RecreationHours', 'TentCampers', 'RVCampers',
       'Backcountry', 'RecreationVisitsTotal', 'RecreationHoursTotal',
       'TentCampersTotal', 'RVCampersTotal', 'BackcountryTotal']
```

```python
#the data values are in string comma number format, getting rid of commas and converting to number
cleaned_data=filtered_data.copy()
for i in value_columns:
    cleaned_data[i]=cleaned_data[i].map(lambda s: int(s.replace(',','')))

#this csv file gives monthly numbers for all parks for all years
cleaned_data.to_csv('../clean_data/cleaned_data_monthly.csv')
```

```python
cleaned_data.head()
```

```python
#further cleaning up the data to include just the yearly total starting in 2001
cleaned_annual_data = cleaned_data.drop_duplicates(subset = ['ParkName','Year'], keep ='last')
cleaned_annual_data = cleaned_annual_data.loc[(cleaned_annual_data['Year']  > 2000),:]
cleaned_annual_data['ParkName']=cleaned_annual_data['ParkName'].map(lambda s: s.replace(' NP',''))
cleaned_annual_data=cleaned_annual_data[['ParkName', 'ParkType', 'Region', 'State', 'Year',
        'RecreationVisitsTotal', 'RecreationHoursTotal',
       'TentCampersTotal', 'RVCampersTotal', 'BackcountryTotal']]
cleaned_annual_data.to_csv('../clean_data/cleaned_annual_data.csv')
cleaned_annual_data  #SAVE THIS
```

```python
#simple calculation summing up visits across all parks 2001-2019
total_ann=cleaned_annual_data.groupby('Year')
total_annual=pd.DataFrame(total_ann['RecreationVisitsTotal'].sum())
total_annual['backcountry_total']=total_ann['BackcountryTotal'].sum()
total_annual['tent_total']=total_ann['TentCampersTotal'].sum()
total_annual=total_annual.reset_index()
total_annual.to_csv('../clean_data/total_annual.csv')
total_annual
```
