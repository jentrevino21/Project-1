```python
#modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
#read csv data
cleaned_annual_data = pd.read_csv('../clean_data/cleaned_annual_data.csv')
total_annual = pd.read_csv('../clean_data/total_annual.csv')
cleaned_annual_data
```

```python
#furthur cleaning data to 
cleaned_annual_data = cleaned_annual_data.loc[(cleaned_annual_data['Year']  > 2009),:]
tops=cleaned_annual_data.groupby('ParkName')
top8=pd.DataFrame(tops['State'].unique())
top8['avg_visits']=tops['RecreationVisitsTotal'].mean()
top8['avg_tent']=pd.DataFrame(tops['TentCampersTotal'].mean())
top8['avg_backcountry']=pd.DataFrame(tops['BackcountryTotal'].mean())
top8.to_csv('../clean_data/top_parks_2010-2019.csv')
top8
```

```python
#want to get the top parks by avg visitors from 2010-2019 by recreational visitors
recvis=top8.sort_values("avg_visits",ascending = False)
recvis=recvis.iloc[:9,:]
recvis=recvis.reset_index()
recvis
```

```python
#top parks by tent campers
tenters=top8.sort_values("avg_tent",ascending = False)
tenters=tenters.iloc[:9,:]
tenters=tenters.reset_index()
tenters
```

```python
#top parks by backcountry campers
backpackers=top8.sort_values("avg_backcountry",ascending = False)
backpackers=backpackers.iloc[:9,:]
backpackers=backpackers.reset_index()
backpackers
```

```python
#function to create bar charts for each type of visit

def barchart(xlegend,yaxis,visit_type):
    plt.figure(figsize=(10,7))
    plt.title(f'National Parks: Annual Average {visit_type} (2010-2019)\n')
    plt.xticks(rotation = 85) 
    plt.ticklabel_format(style='plain')
    plt.ylim(0,yaxis.max()+yaxis.max()*.1)
    
    plt.bar(xlegend,yaxis,width=.8,color='dodgerblue')
    ax = plt.gca()
    ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
    loc,lab=plt.xticks()
    #this formats the y axis to comma number
    for i in range(len(xlegend)):
        plt.text(loc[i], yaxis[i]+yaxis.max()*.02, f'{int(round(yaxis[i],0)):,}',ha='center')
    #plt.tight_layout()
    plt.gcf().subplots_adjust(bottom=0.30)
    plt.savefig(f"../plots/{visit_type}.png")
```

```python
barchart(backpackers['ParkName'],backpackers['avg_backcountry'],'Backcountry Campers')
#loc,lab=plt.xticks()
```

```python
barchart(tenters['ParkName'],tenters['avg_tent'],'Tent Campers')
```

```python
barchart(recvis['ParkName'],recvis['avg_visits'],'Recreational Visitors')
```

```python
state = cleaned_annual_data.drop_duplicates(subset = ['ParkName'], keep ='last')
top_states=pd.DataFrame(state['State'].value_counts())
top_states=top_states.reset_index()
```

```python
top_states=top_states.rename(columns={'index':'State','State':'Count'})
top_5=top_states.iloc[:5,:]
top_5
```

```python
xlab=top_5['State']
yax=top_5['Count']

plt.figure(figsize=(6,6))
plt.yticks(np.arange(0,yax.max()+1,1))
#plt.title(f'National Parks: Annual Average {visit_type} (2010-2019)\n')

  

plt.bar(xlab,yax,width=.6,color='dodgerblue')

#plt.tight_layout()
plt.savefig(f"../plots/state_np_counts_top5.png")
```

```python
xl=top_states['State']
ya=top_states['Count']

plt.figure(figsize=(12,6))
plt.yticks(np.arange(0,yax.max()+1,1))
#plt.title(f'National Parks: Annual Average {visit_type} (2010-2019)\n')

plt.tight_layout(h_pad=0) 

plt.bar(xl,ya,width=.7,color='dodgerblue')



plt.savefig(f"../plots/state_np_counts.png")
plt.show()
```

```python
#graph total visitation numbers across all national parks from 2001-2019

x_axis=total_annual['Year']
y_axis=total_annual['RecreationVisitsTotal']
plt.figure(figsize=(10,7))
plt.xticks(np.arange(2001,2020,1),rotation=90)
plt.yticks(np.arange(round(y_axis.min(),-7)-10000000,round(y_axis.max(),-7)+10000000,2500000))


plt.title('Total Annual Recreational Visits Across All National Parks\n2001-2019\n')
plt.plot(x_axis,y_axis,marker="o", linewidth=1,color='dodgerblue')
ax = plt.gca()
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.annotate('89,815,063',(2016.6,89505063),color='red')
plt.annotate('68,064,224',(2013.5,67864224),color='red')
#plt.annotate('70,168,104',(2000.7,68768104),color='red')
plt.axvline(x=2013,ls='--',color='blue',alpha=.25)

plt.savefig("../plots/total_annual_visits.png")
plt.show()
```

```python
back_x_axis=total_annual['Year']
back_y_axis=total_annual['backcountry_total']
plt.figure(figsize=(10,7))
plt.xticks(np.arange(2001,2020,),rotation=90)

plt.title('Total Backcountry Campers Across All National Parks\n2001-2019\n')


plt.plot(back_x_axis,back_y_axis,marker="o", linewidth=1,color='dodgerblue')

ax = plt.gca()
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.annotate('1,521,510',(2016.45,1521510),color='red')
plt.annotate('1,136,255',(2013.3,1136255),color='red')

plt.yticks(np.arange(950000,1650000,50000))
plt.axvline(x=2013,ls='--',color='blue',alpha=.25)
plt.savefig("../plots/total_backcountry.png")
plt.show()
```

```python
x_tent=total_annual['Year']
y_tent=total_annual['tent_total']
plt.figure(figsize=(10,7))
plt.xticks(np.arange(2001,2020,),rotation=90)
plt.title('Total Tent Campers Across All National Parks\n2001-2019\n')

plt.plot(x_tent,y_tent,marker="o", linewidth=1,color='dodgerblue')

ax = plt.gca()
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.annotate('2,995,209',(2016.5,2985209),color='red')
plt.annotate('2,477,355',(2013.3,2297355),color='red')

plt.yticks(np.arange(2100000,3200000,100000))
plt.axvline(x=2013,ls='--',color='blue',alpha=.25)

plt.savefig("../plots/total_tent.png")
plt.show()
```
